```python
import requests as r
import csv
```

```python
def file_write():
    file_name = input("輸入檔案名稱(含副檔名):")
    file_place = input("輸入檔案儲存位置:")
    file_URL = input("輸入檔案網址:")
    if file_place == "":
        file_place = "/opendata/"
    elif file_place[-1] != "/":
        file_place+="/"
    if file_name[file_name.index(".")::] == ".csv" or file_name[file_name.index(".")::] == ".json":
        with open(file_place + file_name,'w',encoding="utf-8") as read_file:
            read_file.write(r.get(file_URL).text)
            print("!儲存成功!")
```

```python
def print_file(file:list):
    for row in file:
        for col in row:
            print(col,end=",")
        print("\b")

with open("/workspaces/PythonCases/class/arrey/aqx_p_434.csv",encoding="utf-8") as read_file:
    file = list(csv.reader(read_file))
while True:
    funcs = ["讀取指定直列","篩選特定橫排","計數","上一步"]
    for i in range(len(funcs)):
        print(f"[{str(i+1).center(3)}]  {funcs[i]}")
    print("[ 0 ]  關閉程式")
    func = int(input("輸入需要的功能代碼:"))
    if func == 0:
        break
    elif func == 1:
        ind = list(map(int,input("第幾列:").split(",")))
        new_list = [[i[j-1] for j in ind] for i in file]
        print(new_list)
```

```python
funcs = ["讀取網路檔案","檔案讀取"]
while True:
    print("".center(20,"="))
    for i in range(len(funcs)):
        print(f"[{str(i+1).center(3)}]  {funcs[i]}")
    print("[ 0 ]  關閉程式")
    func = int(input("輸入需要的功能代碼:"))
    if func == 0:
        print("!!感謝使用!!")
        break
    elif func == 1:
        file_write()
```
